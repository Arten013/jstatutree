```python
for sample_code in reikiset_path.glob('**/*.xml'):
    sample_code = lawdata.ReikiCode.init_from_path(sample_code)
    try:
        tree = Jstatutree(sample_code.to_path(reikiset_path))
    except LawError:
        continue
    if tree.findall('//Column'):
        for e in tree.iter():
            print(e)
            print(e.text)
```
